```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df_co2_2021_03 = pd.read_csv("csv/82000278_Toamnei_CO2_2021_03.csv")
df_co2_2022_03 = pd.read_csv("csv/82000278_Toamnei_CO2_2022_03.csv")
df_co2_2021_08 = pd.read_csv("csv/82000278_Toamnei_CO2_2021_08.csv")
```

```python
X_co2_2021_03 = df_co2_2021_03[["time"]].values
Y_co2_2021_03 = df_co2_2021_03[["co2"]].values

i_co2_2021_03=0
x_co2_2021_03=[]
y_co2_2021_03=[]
for el in X_co2_2021_03:
    x_co2_2021_03.append(i_co2_2021_03)
    y_co2_2021_03.append(Y_co2_2021_03[i_co2_2021_03][0])
    i_co2_2021_03=i_co2_2021_03+1
```

```python
X_co2_2022_03 = df_co2_2022_03[["time"]].values
Y_co2_2022_03 = df_co2_2022_03[["co2"]].values

i_co2_2022_03=0
x_co2_2022_03=[]
y_co2_2022_03=[]
for el in X_co2_2022_03:
    x_co2_2022_03.append(i_co2_2022_03)
    y_co2_2022_03.append(Y_co2_2022_03[i_co2_2022_03][0])
    i_co2_2022_03=i_co2_2022_03+1
```

```python
X_co2_2021_08 = df_co2_2021_08[["time"]].values
Y_co2_2021_08 = df_co2_2021_08[["co2"]].values

i_co2_2021_08=0
x_co2_2021_08=[]
y_co2_2021_08=[]
for el in X_co2_2021_08:
    x_co2_2021_08.append(i_co2_2021_08)
    y_co2_2021_08.append(Y_co2_2021_08[i_co2_2021_08][0])
    i_co2_2021_08=i_co2_2021_08+1
```

```python
plt.figure(figsize=(30,15))
plt.plot(X_co2_2021_03, y_co2_2021_03, color = 'blue', label="CO2 values - March 2021")
plt.plot(X_co2_2022_03, y_co2_2022_03, color = 'red', label="CO2 values - March 2022")
plt.plot(X_co2_2021_08, y_co2_2021_08, color = 'green', label="CO2 values - August 2021")
plt.title('CO2 values - Toamnei - Different months', fontsize=18)
plt.xlabel('TimeStamp', fontsize=16)
plt.ylabel('Value', fontsize=16)
plt.legend(fontsize=16)
# plt.savefig('plots_from_csv/CO2_Toamnei_2021-07_Forest-GridSearch_1month.jpg', bbox_inches = 'tight')
plt.show()
```

```python
#{bootstrap=True, max_depth=60, max_features='sqrt', min_samples_leaf=4, min_samples_split=6, n_estimators=700} 
from sklearn.ensemble import RandomForestRegressor
regressor_co2_2021_03 = RandomForestRegressor(bootstrap=True, max_depth=60, max_features='sqrt', min_samples_leaf=4, min_samples_split=6, n_estimators=700).fit(X_co2_2021_03, y_co2_2021_03)
y_pred_co2_2021_03 = regressor_co2_2021_03.predict(X_co2_2021_03)
```

```python
#{bootstrap=True, max_depth=60, max_features='sqrt', min_samples_leaf=4, min_samples_split=6, n_estimators=700} 
from sklearn.ensemble import RandomForestRegressor
regressor_co2_2022_03 = RandomForestRegressor(bootstrap=True, max_depth=60, max_features='sqrt', min_samples_leaf=4, min_samples_split=6, n_estimators=700).fit(X_co2_2022_03, y_co2_2022_03)
y_pred_co2_2022_03 = regressor_co2_2022_03.predict(X_co2_2022_03)
```
