```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.isotonic import IsotonicRegression
```

```python
dataframe = pd.read_csv("csv/82000278_Toamnei_PM10_2021_12.csv")
```

```python
X = dataframe[["time"]].values
Y = dataframe[["pm10"]].values
```

```python
i=0
x=[]
y=[]
for el in X:
    x.append(i)
    y.append(Y[i][0])
    i=i+1
```

```python
clf = IsotonicRegression()
clf.fit(x, y)
y_pred = clf.predict(x)
```

```python
plt.figure(figsize=(30,15))
plt.plot(X, y, color='blue', label="PM10 values")
plt.plot(X, y_pred, color='red', label='Isotonic Regression')
plt.title('Isotonic Regression on December 2021 in Toamnei', fontsize=18)
plt.xlabel('TimeStamp', fontsize=16)
plt.ylabel('Value', fontsize=16)
plt.legend(fontsize=16)
plt.savefig('plots_from_csv/PM10_Toamnei_2021-12_Isotonic.jpg', bbox_inches = 'tight')
plt.show()
```
