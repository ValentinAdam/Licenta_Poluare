```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
dataframe = pd.read_csv("csv/82000278_Toamnei_CO2_2021_7_SHORT.csv")
dataframe
```

```python
X = dataframe[["time"]].values
Y = dataframe[["co2"]].values
```

```python
i=0
x=[]
y=[]
for el in X:
    x.append(i)
    y.append(Y[i][0])
    i=i+1
```

```python
#{'bootstrap': True, 'max_depth': 55, 'max_features': 'sqrt', 'min_samples_leaf': 5, 'min_samples_split': 6, 'n_estimators': 600}
from sklearn.ensemble import RandomForestRegressor
clf = RandomForestRegressor(bootstrap=True, max_depth=55, max_features='sqrt', min_samples_leaf=5, min_samples_split=6, n_estimators=600).fit(X, y)
y_pred = clf.predict(X)
```

```python
plt.figure(figsize=(30,15))
plt.plot(X, y, color = 'blue', label="CO2 values")
plt.plot(X, y_pred, color = 'red', label = 'Forest Regression')
# plt.scatter(X, y, color = "green")
# plt.scatter(X, y_pred, color = "black")
plt.title('Forest Regression using Grid Search parameters on a week in July 2021 in Toamnei', fontsize=18)
plt.xlabel('TimeStamp', fontsize=16)
plt.ylabel('Value', fontsize=16)
plt.legend(fontsize=16)
plt.savefig('plots_from_csv/CO2_Toamnei_2021-07_Forest-GridSearch_1week.jpg', bbox_inches = 'tight')
plt.show()
```
