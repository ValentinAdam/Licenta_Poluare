```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor
```

```python
dataframe = pd.read_csv("csv/82000278_Toamnei_PM10_2021_12.csv")
```

```python
X = dataframe[["time"]].values
Y = dataframe[["pm10"]].values
```

```python
i=0
x=[]
y=[]
for el in X:
    x.append(i)
    y.append(Y[i][0])
    i=i+1
```

```python
rng = np.random.RandomState(1)
clf = AdaBoostRegressor( DecisionTreeRegressor(max_depth=4), n_estimators=300, random_state=rng)
clf.fit(X, y)
y_pred = clf.predict(X)
```

```python
plt.figure(figsize=(30,15))
plt.plot(X, y, color='blue', label="PM10 values")
plt.plot(X, y_pred, color='red', label='Decision Tree')
plt.title('Decision Tree regression on December 2021 in Toamnei', fontsize=18)
plt.xlabel('TimeStamp', fontsize=16)
plt.ylabel('Value', fontsize=16)
plt.legend(fontsize=16)
plt.savefig('plots_from_csv/PM10_Toamnei_2021-12_Decision-Tree.jpg', bbox_inches = 'tight')
plt.show()
```
